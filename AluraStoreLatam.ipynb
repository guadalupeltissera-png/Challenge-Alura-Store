### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

#1. Análisis de facturación



```python

suma_precio_tienda =tienda['Precio'].sum()
print(f'El ingreso total de la primera tienda es de: ${suma_precio_tienda}')

suma_precio_tienda2 =tienda2['Precio'].sum()
print(f'El ingreso total de la segunda tienda es de: ${suma_precio_tienda2}')

suma_precio_tienda3 =tienda3['Precio'].sum()
print(f'El ingreso total de la segunda tienda es de: ${suma_precio_tienda3}')

suma_precio_tienda4 = tienda4['Precio'].sum()
print(f'El ingreso total de la segunda tienda es de: ${suma_precio_tienda4}')
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
suma_precio = [suma_precio_tienda, suma_precio_tienda2, suma_precio_tienda3, suma_precio_tienda4]

plt.figure(figsize=(8,5))

plt.plot(suma_precio, tiendas, 'o', markersize=10, color='purple')

plt.xlabel('Ingreso total en miles de millones ($)')
plt.ylabel('Tiendas')
plt.title('Comparación del ingreso total de cada tienda')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_por_producto_tienda = tienda.groupby('Producto')['Producto'].count().reset_index(name='CantidadVendida')
mas_ventas_tienda =  ventas_por_producto_tienda.nlargest(5,'CantidadVendida').sort_values('CantidadVendida')
menos_ventas_tienda =  ventas_por_producto_tienda.nsmallest(5,'CantidadVendida').sort_values('CantidadVendida')
mas_menos_ventas_tienda = pd.concat([menos_ventas_tienda,mas_ventas_tienda])
```

```python
colores = ['green' if x in mas_ventas_tienda['Producto'].values else 'red' for x in mas_menos_ventas_tienda['Producto']]

plt.figure(figsize=(8,4))
plt.barh(mas_menos_ventas_tienda['Producto'], mas_menos_ventas_tienda['CantidadVendida'], color=colores)
plt.title('Productos más y menos vendidos - Tienda 1')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.show()
```

```python
ventas_por_producto_tienda2 = tienda2.groupby('Producto')['Producto'].count().reset_index(name='CantidadVendida')
mas_ventas_tienda2 =  ventas_por_producto_tienda2.nlargest(5,'CantidadVendida').sort_values('CantidadVendida')
menos_ventas_tienda2 =  ventas_por_producto_tienda2.nsmallest(5,'CantidadVendida').sort_values('CantidadVendida')
mas_menos_ventas_tienda2 = pd.concat([menos_ventas_tienda2,mas_ventas_tienda2])

colores = ['green' if x in mas_ventas_tienda2['Producto'].values else 'red' for x in mas_menos_ventas_tienda2['Producto']]

plt.figure(figsize=(8,4))
plt.barh(mas_menos_ventas_tienda2['Producto'], mas_menos_ventas_tienda2['CantidadVendida'], color=colores)
plt.title('Productos más y menos vendidos - Tienda 2')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.show()
```

```python
ventas_por_producto_tienda3= tienda3.groupby('Producto')['Producto'].count().reset_index(name='CantidadVendida')
mas_ventas_tienda3 =  ventas_por_producto_tienda3.nlargest(5,'CantidadVendida').sort_values('CantidadVendida')
menos_ventas_tienda3 =  ventas_por_producto_tienda3.nsmallest(5,'CantidadVendida').sort_values('CantidadVendida')
mas_menos_ventas_tienda3 = pd.concat([menos_ventas_tienda3,mas_ventas_tienda3])

colores = ['green' if x in mas_ventas_tienda3['Producto'].values else 'red' for x in mas_menos_ventas_tienda3['Producto']]

plt.figure(figsize=(8,4))
plt.barh(mas_menos_ventas_tienda3['Producto'], mas_menos_ventas_tienda3['CantidadVendida'], color=colores)
plt.title('Productos más y menos vendidos - Tienda 3')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.show()
```

```python
ventas_por_producto_tienda4= tienda4.groupby('Producto')['Producto'].count().reset_index(name='CantidadVendida')
mas_ventas_tienda4 =  ventas_por_producto_tienda4.nlargest(5,'CantidadVendida').sort_values('CantidadVendida')
menos_ventas_tienda4 =  ventas_por_producto_tienda4.nsmallest(5,'CantidadVendida').sort_values('CantidadVendida')
mas_menos_ventas_tienda4 = pd.concat([menos_ventas_tienda4,mas_ventas_tienda4])

colores = ['green' if x in mas_ventas_tienda4['Producto'].values else 'red' for x in mas_menos_ventas_tienda4['Producto']]

plt.figure(figsize=(8,4))
plt.barh(mas_menos_ventas_tienda4['Producto'], mas_menos_ventas_tienda4['CantidadVendida'], color=colores)
plt.title('Productos más y menos vendidos - Tienda 4')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.show()
```

# 5. Envío promedio por tienda

```python
costo_envio_tienda = round(tienda['Costo de envío'].mean(),2)
print(f'El costo de envio promedio en la primera tienda es de: ${costo_envio_tienda}')
```

```python
costo_envio_tienda2 = round(tienda2['Costo de envío'].mean(),2)
print(f'El costo de envio promedio en la segunda tienda es de: ${costo_envio_tienda2}')
```

```python
costo_envio_tienda3 = round(tienda3['Costo de envío'].mean(),2)
print(f'El costo de envio promedio en la tercera tienda es de: ${costo_envio_tienda3}')
```

```python
costo_envio_tienda4 = round(tienda4['Costo de envío'].mean(),2)
print(f'El costo de envio promedio en la cuarta tienda es de: ${costo_envio_tienda4}')
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios = [costo_envio_tienda, costo_envio_tienda2, costo_envio_tienda3, costo_envio_tienda4]

plt.style.use('_mpl-gallery')

x = np.arange(len(tiendas))
markerline, stemlines, baseline = plt.stem(x, envios, basefmt=" ", linefmt='lightpink', markerfmt='hotpink')
plt.xticks(x, tiendas)
plt.ylabel('Costo de envío ($)')
plt.title('Comparacion del costo de envío promedio entre las tiendas')
plt.show()
```

Luego de analizar el ingreso total, la distribución de ventas por categoría, la calificación promedio de cada tienda, el desempeño de los productos más y menos vendidos y el costo de envío llegué a la conclusión de que la tienda a cerrar es la n°4.
Esto se debe a que a pesar de mantener un volumen de ventas similar al de las otras tiendas en ciertas categorías, su ingreso total es menor en comparación.
Por lo tanto, recomiendo estrategicamente el cierre de la cuarta tienda ya que su desempeño financiero es menor al de las otras.
